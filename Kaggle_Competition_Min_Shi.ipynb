# <font color = 'pickle'> Data Loader

```python
df = pd.read_csv(data_folder/'train.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.iloc[0:10, ]
```

```python
# def get_x(df): return str(small_train_images) + '/' + (df['image'])
def get_x(df):
  return data_folder/'train/train'/df['image']
def get_y(df):
  return df['labels'].split(' ')
dblock = DataBlock(get_x= get_x,
                   get_y = get_y,
                   splitter = RandomSplitter(seed = 42))
dsets = dblock.datasets(df)
dsets.train[15]
```

```python
dblock = DataBlock(blocks= (ImageBlock, MultiCategoryBlock),
                   get_x = get_x,
                   get_y = get_y,
                  splitter = RandomSplitter(seed = 42))
dsets = dblock.datasets(df)
dsets.train[15]
```

```python
dsets.train.vocab
```

```python
idxs = torch.where(dsets.train[15][1] == 1.)[0]
```

```python
dsets.train.vocab[idxs]
```

```python
dblock = DataBlock(blocks= (ImageBlock, MultiCategoryBlock),
                  get_x= get_x,
                  get_y = get_y,
                  splitter = RandomSplitter(seed = 42),
                  item_tfms= Resize(400, method = 'squish'),
                  batch_tfms= aug_transforms(size = 256, min_scale = 0.75))
dsets = dblock.datasets(df)
dsets.train[3]
```

```python
import os
os.cpu_count()
```

```python
dls = dblock.dataloaders(df, bs = 64, num_workers =8)
```

```python
dls.train.show_batch(max_n=3)
```

```python
dls.device
```

# <font color = 'pickle'> Model

```python
f1score = F1ScoreMulti(thresh = 0.5) # F1 socre is calculated based on precision and recall
loss_func = nn.BCEWithLogitsLoss() # binary cross entropy loss
# loss_func = nn.CrossEntropyLoss() # multi class cross entropy loss
learn = vision_learner(dls,'convnext_tiny', metrics = [F1ScoreMulti(thresh = 0.5)], path = model_folder).to_fp16()
```

```python
learn.lr_find(suggest_funcs = (minimum, steep, valley, slide))
```

```python
learn.fine_tune(3, 0.02512) # 3 epoches
```
